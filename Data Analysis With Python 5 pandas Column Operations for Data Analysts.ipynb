```python
import pandas as pd

names = ["Robert", "Joseph", "David", "Kasie", "Derick"]
age = [23, 34, 21, 55, 49]
profession = ["Software engineer", "Data Analyst", 
              "Data Scientist", "Doctor", "Actor"]
salary = [60000, 75000, 85000, 80000, 200000]


df = pd.DataFrame({"Names": names, "Age":age, "Profession":profession,
                   "Salary": salary})
df
```

```python
# Renaming the "profession" column
df.rename(columns={"Profession":"Job_Title"}, inplace=True)
df
```

```python
# Mapper object to apply to the axis
mapper = {"Names": "First_Name", "Salary": "Salary_Per_Year"}

df.rename(mapper=mapper, axis=1, inplace=True)
df
```

```python
# Specifying the desired order of columns
desired_order = ['First_Name', 'Job_Title', 'Salary_Per_Year', 'Age']

# Passing the desired order to the DataFrame
df = df[desired_order]
df
```

```python
# Set 'Name' and 'City' columns as the new index
df.set_index(["First_Name", "Age"], inplace=True)
df
```

```python
# Reset the MultiIndex and move it back to columns
df.reset_index(inplace=True)
df
```

```python
# Adding multiple columns
df = df.assign(Favorite_Car=["Toyota","Mercedes-Benz", "BMW", "Honda", "Ford"], 
               Favorite_Fruit =["Orange", "Apple", "Mango", "Pear", "GrapeFruit"])

df
```

```python
# Specify the columns to be dropped
columns_to_drop = ['First_Name', 'Age']

# Droping the specified columns
df.drop(columns=columns_to_drop, inplace=True)
df
```
